```python
class Contact:
    def __init__(self, name, address, phone, email):
        self.name = name
        self.address = address
        self.phone = phone
        self.email = email
        
        
    def matches(self, searchstring):
        if self.name.lower().find(searchstring.lower()) != -1:
            return True
        if self.address.lower().find(searchstring.lower()) != -1:
            return True
        if self.phone.lower().find(searchstring.lower()) != -1:
            return True
        if self.email.lower().find(searchstring.lower()) != -1:
            return True
        return False
        
    def print(self):
        print("---------------------------")
        print(" ", self.name)
        print(" ", self.address)
        print(" ", self.phone)
        print(" ", self.email)
        print("===========================")
        
    def update(self):
        self.print()
        print("which field do you want to update?")
        up = input("[n]ame, [a]ddress, [p]hone, [e]mail: ")
        if up == 'n':
            self.name = input("new name: ")
        elif up == 'a':
            self.address = input("new address: ")
        elif up == 'p':
            self.phone = input("new phone number: ")
        elif up == 'e':
            self.email = input("new email address: ")
        else:
            print("invalid field, no update.")
        
        
def new_contact():
    name=input("name: ")
    address = input("address: ")
    phone = input("phone: ")
    email = input("email: ")
    return Contact(name, address, phone, email)
```

```python
class Book:
    def __init__(self):
        self.contacts = []
        
    def add(self, contact):
        self.contacts.append(contact)
        
    def search(self, searchstring):
        result = []
        for c in self.contacts:
            if c.matches(searchstring):
                result.append(c)
        return result
    
    def update(self, index):
        to_update = self.contacts[index]
        to_update.update()
        
    def delete(self, index):
        new_contacts = []
        for c in range(len(self.contacts)):
            if c != index:
                new_contacts.append(self.contacts[c])
        self.contacts = new_contacts
        
    def print(self):
        print("Content:")
        for c in range(len(self.contacts)):
            print(f"contact #{c}:")
            self.contacts[c].print()
```

```python
book = Book()
```

```python
johnny = Contact("Johnny", "Johnstown", "123456", "johnny@abcd.com")
susan = Contact("Susan", "Susanville", "54321", "suzie@efgh.com")
harry = Contact("Harald", "Home", "911", "harryh@gmx.at")
```

```python
book.add(johnny)
book.add(susan)
book.add(harry)
```

```python
def run_addressbook():
    running = True
    while running:
        print("1) create a new contact")
        print("2) print contacts")
        print("3) search for contact")
        print("4) update contact details")
        print("5) delete contact")
        print("q) quit")
        choice = input("> ")
        if choice == "1":
            c = new_contact()
            book.add(c)
        elif choice == "2":
            book.print()
        elif choice == "3":
            search = input("search term: ")
            found = book.search(search)
            print("found following contacts:")
            for f in found:
                f.print()
        elif choice == '4':
            try:
                book.print()
                print("which contact do you want to update?")
                to_update = int(input("enter number (c=cancel):"))
                book.update(to_update)
            except:
                print("invalid input, not updated!")
            
        elif choice == '5':
            try:
                book.print()
                print("which contact do you want to delete?")
                to_delete = int(input("enter number (c=cancel):"))
                book.delete(to_delete)
            except:
                print("invalid input, not deleted!")
        elif choice == "q":
            print("thank you for using the address book")
            print("goodbye!")
            running = False
        
```

```python
run_addressbook()
```
